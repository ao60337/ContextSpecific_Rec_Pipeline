```python
import cobra
import pandas as pd
```

```python
generic_model = cobra.io.read_sbml_model('.../reconstruction_pipeline/utils/human1_models/HumanGEM_Consistent_COVID19')
generic_model
```

```python
path = 'results/HumanGEM_Lung_COVID19_local2_1_3_2_fastcore_t0_msc.csv'

mcs = pd.read_csv(path, index_col=0).T
mcs
```

```python
mcs_dict = mcs.to_dict(orient='list')
mcs_dict
```

```python
wt_simulation = generic_model.optimize()
generic_model.summary()
```

```python
new_mcs = {}
i = 0

for cut_set in mcs_dict.keys():
    with generic_model:
        generic_model.reactions.get_by_id(mcs_dict[cut_set][0]).knock_out()
        generic_model.reactions.get_by_id(mcs_dict[cut_set][1]).knock_out()
        fba = generic_model.optimize()
        if fba.objective_value >= (0.001 * wt_simulation.objective_value):
            new_mcs[i] = [mcs_dict[cut_set][0], mcs_dict[cut_set][1], fba.objective_value]
            i += 1

new_mcs
```

```python
new_mcs['index'] = ['Reaction 1', 'Reaction 2', 'fba_simulation']
new_mcs
```

```python
df = pd.DataFrame.from_dict(new_mcs).set_index('index').T
df
```

```python
res_path = 'Results/HumanGEM_Lung_COVID19_local2_1_3_2_fastcore_t0_msc_filtered.csv'

df.to_csv(res_path)
```
